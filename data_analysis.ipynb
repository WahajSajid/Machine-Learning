```python
#Import Libraries
import pandas as pd
import numpy as np

data_read = pd.read_excel("C:/Users/Wahaj Sajid/Downloads/University Students Data.xlsx")

data = pd.DataFrame(data_read)
data
```

```python
data.drop(index=0)
```

```python



acutal_value = "Computer Sciences"
departmet_changes = {"Bscs": acutal_value, "BS Computer science" :acutal_value, "Computer science": acutal_value, "BSCS":acutal_value, "CS":acutal_value, "Bs cs": acutal_value, "Bs Computer Science": acutal_value, "Cs" :acutal_value, "B's Computer Science":acutal_value}
semester_changes = {"6th": "6", "6th Semester": "6", "4th":"4", "4th Semester":"4"}
data['Department'] = data["Department"].replace(departmet_changes)
data["Current Semester"] = data["Current Semester"].replace(semester_changes)
```

```python
data['Current CGPA'] = pd.to_numeric(data['Current CGPA'], errors='coerce')
```

```python
#Imputing null values
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
data.loc[2, "Current CGPA"] = 3.3
data.loc[1, "Current Semester"] = 4
data.loc[4,"Current Semester"] = 6
data.loc[13,"Current Semester"] = 4
data['Current CGPA'] = imputer.fit_transform(data[['Current CGPA']])
data['Current CGPA'].sort_values(ascending=False)
data.drop(index=0)
```
