```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#here x is the input (feature) and y is the output (target)
x_train = np.array([1.0,2.0])
y_train = np.array([300.0,500.0])
print(f"x_train:{x_train}")
print(f"y_train:{y_train}")


#get the size of the training data
m = x_train.shape[0]
print(f"shape of the data: {m}")


#get the ith training example of data
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"x_i: {x_i} and y_i: {y_i}")
```

```python
#PLot the data
plt.scatter(x_train, y_train, marker = 'x', c = 'r')
plt.title('House Prices')
plt.ylabel('Prices of houses')
plt.xlabel("Size of houses")
plt.show()
```

```python
def compute_output(x,w,b):
    m = x.shape[0]
    f = np.zeros(m)
    for i in range(m):
        f[i] = w * x[i] +b
    return f
```

```python
w = 200
b = 100
prediction = compute_output(x_train, w,b)

#plot the prediction
plt.plot(x_train, prediction, color='blue', label='Prediction')
plt.scatter(x_train, y_train, marker='x', c='r', label='Training Data')
plt.ylabel("Prices of houses")
plt.xlabel("Size of houses")
plt.title('House Prices with Prediction')
plt.legend()
plt.show()
```
