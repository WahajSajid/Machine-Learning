```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load titanic dataset
data = sns.load_dataset('titanic')

# Visualize the data
plt.figure(figsize=(8, 5))
sns.heatmap(data.isnull(), cbar=False)
plt.show()
```

```python
#Load the data
data = sns.load_dataset("titanic")
data.head()
```

```python
#getting the null values present in the data
data.isnull().sum().sort_values(ascending=False)
```

```python
#getting the null values present in the data
data.isnull().sum().sort_values(ascending=False)
```

```python
#getting the percentage of null values present in the each column
round(data.isnull().sum().sort_values(ascending=False) / len(data) *  100, 2)
```

Imputing Values using Mean Median and Mode. We use Mean and Median for the numerical data to impute. We use Mode for the categorical data to impute

```python
#getting the mean of age to impute the missing values
#Replacing the missing values of age in the data using median
data["age"]  = data["age"].fillna(data["age"].median)
```

```python
#getting the null values present in the data
data.isnull().sum().sort_values(ascending= False)
```

```python
#droping the deck column from the data
data.drop("deck", axis=1, inplace = True)
```

```python
# Replacing the missing values of embark_town with Mode as it is a categorical variable
data["embark_town"]  = data['embark_town'].fillna(data['embark_town'].mode()[0])
data["embarked"]  = data['embarked'].fillna(data['embarked'].mode()[0])
```

We can also Impute using SK learn

```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.impute import SimpleImputer

loadData = sns.load_dataset("titanic")
loadData.head()
```

```python
#imputing the missing values in the data using Simple Imputer of the sk learn
imputer = SimpleImputer(strategy="mean")
loadData['age'] = imputer.fit_transform(loadData[['age']])
```

```python
loadData.isnull().sum().sort_values(ascending=False)
```

Imputing values using Multivariate technique

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

#imputing the missing values using iterative imputer 
imputer = IterativeImputer(n_nearest_features=6, max_iter=20)
loadData['age'] = imputer.fit_transform(loadData[['age']])
```

```python
loadData.isnull().sum().sort_values(ascending=False)
```

Imputing Values using forward and backward fill

```python
#Imputing missing values using forward fill
loadData['age'] = data['age'].ffill()

#Imputind missing values using backward fill
loadData['embarked'] = data['embarked'].bfill()
```

```python
loadData.isnull().sum().sort_values(ascending=False)
```

Imputing missing values in the data using KNN Imputer

```python
from sklearn.impute import KNNImputer

#imputing missing values using KNN Imputer
imputer = KNNImputer(n_neighbors = 5)
loadData['age'] = imputer.fit_transform(loadData[['age']])
```

```python
loadData.isnull().sum().sort_values(ascending=False)
```

We can also drop all the missing values from the data

```python
#it will remove all the missing values from the data
loadData.dropna(inplace=True)
```
