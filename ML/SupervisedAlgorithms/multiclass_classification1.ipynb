# Classification of Flowers (using iris dataset) the algorithm used is Multiclass Classification Neural Network (Softmax)

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import pandas as pd
from tensorflow.keras.losses import SparseCategoricalCrossentropy
```

```python
#load the dataset
iris_data = sns.load_dataset('iris')
data = sns.load_dataset('iris')
```

```python
data.tail(70)
```

```python
iris_data.tail(70)
```

```python
#encode the data
encoder = LabelEncoder()
iris_data['species'] = encoder.fit_transform(iris_data[['species']])
```

```python
#defining x and y
x = iris_data.drop('species', axis=1)
y = iris_data['species']
```

```python
#train test split the model 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.8)
```

```python
#train the model
model = Sequential([
    Dense(units=30, activation='relu'),
    Dense(units=10, activation='relu'),
    Dense(units=3, activation='softmax')
])
model.compile(loss = SparseCategoricalCrossentropy(),optimizer = 'adam')
model.fit(x_train, y_train, epochs = 50, batch_size = 25)
```

```python
#test the model
test_features = np.array([[6.7,3.0,5.2 ,2.5]])
logits = model(test_features)
prediction = tf.nn.softmax(logits)
print(prediction)
```

```python
#test the model
test_features = np.array([[6.7,3.0,5.2 ,2.5]])

predict = model.predict(test_features)

prediction = np.argmax(predict, axis = 1)

decoded_prediction = encoder.inverse_transform(prediction)
print(decoded_prediction[0])
```

```python
#evaluate the model
loss = model.evaluate(x_test, y_test)
print(loss)
```
