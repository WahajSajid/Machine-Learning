# L1 and L2 Normalizer 

```python
# import libraries
import numpy as np
import pandas as pd 
from sklearn.preprocessing import Normalizer, QuantileTransformer
import matplotlib.pyplot as plt
```

```python
data = [[4, 1, 2, 2],
        [1, 3, 9, 3],
        [5, 7, 5, 1]]
normalizer = Normalizer(norm= "l2")
print(normalizer.fit_transform(data))
```

# Log Transformation 

```python
import pandas as pd
import numpy as np  
import seaborn as sns
data1 = {"Values":[5, 10, 20 , 30 , 100, 200, 500,  700, 900]}
dataFrame = pd.DataFrame(data1)  
sns.histplot(dataFrame, x = "Values", kde=True)
```

```python
dataFrame["log_values"] = np.log(dataFrame["Values"])
dataFrame
```

```python
sns.histplot(dataFrame, x = "log_values", kde=True)
```

```python
# standardization the data
qt = QuantileTransformer(output_distribution= "normal")
dataFrame["quantile_transformer"] = qt.fit_transform(dataFrame[["Values"]])
dataFrame
```

```python
for column in dataFrame.columns:
    sns.histplot(dataFrame, x = column, kde=True)
    plt.show()
```
