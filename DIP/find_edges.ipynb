```python
#import libraries
import cv2
import matplotlib.pyplot as plt
```

```python
#implementing the image edge finding functionality

# Read the image
image = cv2.imread(r"C:\Users\Wahaj Sajid\Desktop\profile image.jpg", cv2.IMREAD_COLOR_RGB)

# Apply Canny edge detection
edges = cv2.Canny(image, 100, 150)

# Show the original and edge-detected images
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Edges')

plt.show()
```
