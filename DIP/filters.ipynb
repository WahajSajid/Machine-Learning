```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


# Read image as grayscale
img = cv2.imread('home_made_food.jpg', cv2.IMREAD_COLOR)

# Gaussian filter
gaussian_blur = cv2.GaussianBlur(img, (5, 5), sigmaX=1.0)  # (kernel size, sigma)

# Display results
plt.figure(figsize=(10, 8))
plt.subplot(2, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off') 
plt.subplot(2, 3, 2)
plt.imshow(gaussian_blur, cmap='gray')
plt.title('Gaussian Blur')
plt.axis('off') 


plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


# Read image as grayscale
img = cv2.imread('home_made_food.jpg', cv2.IMREAD_GRAYSCALE)

# Smoothing filter (mean filter)
mean_blur = cv2.blur(img, (5, 5))  # (kernel size)

# display results
plt.figure(figsize=(10, 8))
plt.subplot(2, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(2, 3, 2)
plt.imshow(mean_blur, cmap='gray')
plt.title('Mean  Blur')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


# Read image as grayscale
img = cv2.imread('home_made_food.jpg', cv2.IMREAD_GRAYSCALE)


# Median filter
median_blur = cv2.medianBlur(img, 5)  # kernel size (must be odd)

# display results
plt.figure(figsize=(10, 8))
plt.subplot(2, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(2, 3, 2)
plt.imshow(median_blur, cmap='gray')
plt.title('Median  Blur')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


# Read image as grayscale
img = cv2.imread('home_made_food.jpg', cv2.IMREAD_GRAYSCALE)


# Bilateral filter
bilateral_blur = cv2.bilateralFilter(img, d=9, sigmaColor=75, sigmaSpace=75)
# d: diameter of pixel neighborhood
# sigmaColor: filter sigma in color space
# sigmaSpace: filter sigma in coordinate space

# display results
plt.figure(figsize=(10, 8))
plt.subplot(2, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(2, 3, 2)
plt.imshow(bilateral_blur, cmap='gray')
plt.title('Bilateral  Blur')
plt.axis('off')
plt.show()
```
