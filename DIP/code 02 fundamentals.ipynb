```python
import cv2
# Path of the image to read
path = r'F:\img\einstein.jpg'
# Read image using OpenCV function imread
img = cv2.imread(path)
# write the image to file using imwrite() function
cv2.imwrite(r'F:\img\out.jpg',img)
```

```python
# add/subtract a scalar number to image (2D array)  
img2 = cv2.add(img[:,:,0],-80)
# img2[0:50,0:50]
```

```python
fig, axs = plt.subplots(1,3,figsize=(12,8))
# fig.suptitle('Brightness changes')
axs[0].imshow(img,cmap='gray')
axs[0].axis('off')
axs[0].set_title('Original Image',fontsize='medium')

axs[1].imshow(img1,cmap='gray')
axs[1].axis('off')
axs[1].set_title('Brighter Image',fontsize='medium')

axs[2].imshow(img2,cmap='gray')
axs[2].axis('off')
axs[2].set_title('Darker Image',fontsize='medium')

plt.show()
```

### Weber ratio


```python
# image with dark background
import numpy as np
img = np.zeros((300,300), dtype=np.uint8)
delta = 100
img[100:200,100:200] = img[100:200,100:200] + delta
cv2.imshow('Window1',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
# img
```

```python
# image with dark background
img = np.zeros((300,300), dtype=np.uint8)+255
delta = -20
img[100:200,100:200] = img[100:200,100:200] + delta
cv2.imshow('Window2',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# image with dark background
import numpy as np
img1 = np.zeros((300,300), dtype=np.uint8) 
img2 = np.zeros((300,300), dtype=np.uint8)+100
img3 = np.zeros((300,300), dtype=np.uint8)+255

img1[100:200,100:200] = 150
img2[100:200,100:200] = 150
img3[100:200,100:200] = 150

cv2.imshow('Dark Image',img1)
cv2.imshow('Gray Image',img2)
cv2.imshow('Bright Image',img3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
path = r'F:\img\lena.jpg'
img = cv2.imread(path, cv2.IMREAD_COLOR) # default opencv read channels in BGR format
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis('off')
plt.show()

img.shape
```

```python
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]
fig, axs = plt.subplots(1,3,figsize=(12,8))
# fig.suptitle('Brightness changes')
axs[0].imshow(R,cmap='gray')
axs[0].axis('off')
axs[0].set_title('R Channel',fontsize='medium')

axs[1].imshow(G,cmap='gray')
axs[1].axis('off')
axs[1].set_title('G Channel',fontsize='medium')

axs[2].imshow(B,cmap='gray')
axs[2].axis('off')
axs[2].set_title('B Channel',fontsize='medium')

plt.show()

R.shape, G.shape, B.shape
```

```python
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]

fig, axs = plt.subplots(1,3,figsize=(12,8))
# fig.suptitle('Brightness changes')
Rnew = np.zeros((R.shape[0],R.shape[1],3), np.uint8)
Gnew = np.zeros((R.shape[0],R.shape[1],3), np.uint8)
Bnew = np.zeros((R.shape[0],R.shape[1],3), np.uint8)

Rnew[:,:,0] = R
Gnew[:,:,1] = G
Bnew[:,:,2] = B

axs[0].imshow(Rnew)
axs[0].axis('off')
axs[0].set_title('R Channel',fontsize='medium')

axs[1].imshow(Gnew)
axs[1].axis('off')
axs[1].set_title('G Channel',fontsize='medium')

axs[2].imshow(Bnew)
axs[2].axis('off')
axs[2].set_title('B Channel',fontsize='medium')

plt.show()
```
