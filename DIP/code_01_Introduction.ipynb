```python
def displayImg(img):
    channels = len(img.shape)
    rgb = img
    if(channels==3):
        rgb =  cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    if(channels<3):
        plt.imshow(rgb, cmap='gray')
    else:
        plt.imshow(rgb)
    plt.axis('off')
    plt.show()
```
