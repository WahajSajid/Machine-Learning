```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def imshowcv(img,title='image'):
    cv2.imshow(title, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def imshowmat(img,title='image'):
    channels = len(img.shape)
    out = img.copy()
    if(channels==3):
        out =  cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    if(channels<3):
        plt.imshow(out, cmap='gray')
    else:
        plt.imshow(out)
    plt.axis('off')
    plt.title(title)
    plt.show()
   
```

### Thresholding

```python
path = r'F:\img\coins.jpg'
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

# threshold, img1 = cv2.threshold(img, 80, 255, cv2.THRESH_BINARY)
threshold, img1 = cv2.threshold(img, 80, 255, cv2.THRESH_BINARY_INV)
# threshold, img1 = cv2.threshold(img, 150, 255, cv2.THRESH_TRUNC)
# threshold, img1 = cv2.threshold(img, 80, 255, cv2.THRESH_TOZERO)
# threshold, img1 = cv2.threshold(img, 80, 255, cv2.THRESH_TOZERO_INV)
# threshold, img1 = cv2.threshold(img, 80, 255, cv2.THRESH_BINARY +cv2.THRESH_OTSU)
# threshold, img1 = cv2.threshold(img, 80, 255, cv2.THRESH_BINARY +cv2.THRESH_TRIANGLE)

fig, axs = plt.subplots(1,2,figsize=(8,8))
# fig.suptitle('Brightness changes')
axs[0].imshow(img,cmap='gray')
axs[0].axis('off')
axs[0].set_title('Original Image',fontsize='medium')

axs[1].imshow(img1,cmap='gray')
axs[1].axis('off')
axs[1].set_title('Threshold Image',fontsize='medium')
plt.show()

print('Threshold value:' + str(threshold))
```

### Log Transofmration

```python
path = r'F:\img\coins.jpg'
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
c=10
# c = 255 / np.log(1 + np.max(img))
img1 = c * (np.log(1.1+img ))

# norm_image = cv2.normalize(img, None, alpha = 0, beta = 1, norm_type = cv2.NORM_MINMAX, dtype = cv2.CV_32F)
# img1 = (np.log(1+norm_image ))
img2 = np.array(img1, dtype = np.uint8)

fig, axs = plt.subplots(1,2,figsize=(8,8))
# fig.suptitle('Brightness changes')
axs[0].imshow(img,cmap='gray')
axs[0].axis('off')
axs[0].set_title('Original Image',fontsize='medium')

axs[1].imshow(img2,cmap='gray')
axs[1].axis('off')
axs[1].set_title('Log Transform',fontsize='medium')
plt.show()

img2.min(), img2.max()
```

```python
# Power Law
path = r'F:\DIP_Images\DIP3E_CH03\Fig0308(a)(fractured_spine).tif'
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
c=1
gamma =0.4
img1 = c * (img**gamma)
# img1 = np.array(img1, dtype = np.uint8)

fig, axs = plt.subplots(1,2,figsize=(8,8))
# fig.suptitle('Brightness changes')
axs[0].imshow(img,cmap='gray')
axs[0].axis('off')
axs[0].set_title('Original Image',fontsize='medium')

axs[1].imshow(img1,cmap='gray')
axs[1].axis('off')
axs[1].set_title('Power Law',fontsize='medium')
plt.show()
```
