# Linear Regression
Linear Regression is a supervised learning algorithm which predicts continuos value and has a constant slope. It's used to predict values within a continuos range for example (sales, price, etc) rather than trying to classify them into categories for example (dog, cats). There are two main types. 
1) Simple Linear Regression - (Only one independent variable x e.g: price)
2)  Multi Linear Regression - (Multiple independent variables x e.g: price, sales etc)

```python
#import libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#load the data
data = sns.load_dataset('tips')
dataFrame = pd.DataFrame(data)
dataFrame.head()
```

```python
#scaling the data
# scaler = StandardScaler()
# dataFrame['total_bill'] = scaler.fit_transform(dataFrame[['total_bill']])
```

```python
#split the data into feature and target
feature = dataFrame[['total_bill']]
target = dataFrame['tip']
```

```python
#split the data into training and testing
feature_train, feature_test, target_train, target_test = train_test_split(feature, target, test_size=0.8)
```

```python
linearModel = LinearRegression()
linearModel.fit(feature_train, target_train)
```

```python
predict = linearModel.predict([[60]])
predict
```

```python
#evaluate the model
print("Mean Squared Error: ", mean_squared_error(target_test, linearModel.predict(feature_test)))
print("R2 Score: ", r2_score(target_test, linearModel.predict(feature_test)))
print('RMSE: ', np.sqrt(mean_squared_error(target_test, linearModel.predict(feature_test))))
```

```python
plt.scatter(feature_test, target_test)
plt.plot(feature_test, linearModel.predict(feature_test), color='red')
plt.show()
```
